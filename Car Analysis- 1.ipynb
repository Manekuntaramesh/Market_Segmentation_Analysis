```python
import numpy as np
import pandas as pd
```

```python
data1 = pd.read_csv(r"C:\rock\Data Science\Fenny\Car details v3.csv")
```

```python
data1
```

```python
# Data Preparation 
```

```python
# missing values
```

```python
data1.isna().sum()
```

```python
# Extract numerical values from strings.
```

```python
data1['mileage'] = data1['mileage'].str.extract('(\d+\.\d+|\d+)').astype(float)
```

```python
data1['engine'] = data1['engine'].str.extract('(\d+\.\d+|\d+)').astype(float)
```

```python
data1['max_power'] = data1['max_power'].str.extract('(\d+\.\d+|\d+)').astype(float)
```

```python
# filling the missing value -- fillna method 
```

```python
data1['mileage'].fillna(data1['mileage'].median(),inplace = True)
data1['engine'].fillna(data1['engine'].median(),inplace = True)
data1['max_power'].fillna(data1['max_power'].median(),inplace = True)
```

```python
data1.isna().sum()
```

```python
data1['torque'] = data1['torque'].str.extract('(\d+\.\d+|\d+)').astype(float)
```

```python
data1['torque'].fillna(data1['torque'].median(),inplace = True)
```

```python
data1['seats'].fillna(data1['seats'].mode()[0],inplace = True)
```

```python
data1.isna().sum()
```

```python
data1.info()
```

```python
data1.head(4)
```

```python
# checking duplicates
```

```python
data1.duplicated().sum()
```

```python
data1.drop_duplicates(inplace = True)
```

```python
data1.duplicated().sum()
```

```python
#clean 'owner' column for simplified analysis
```

```python
data1['ower_clean'] = data1['owner'].apply(lambda x:x.split()[0])
```

```python
# Extract brand from car name
```

```python
data1['brand'] = data1['name'].apply(lambda x: x.split()[0])
```

```python
from datetime import datetime
```

```python
current_year = datetime.now().year
data1['car_age'] = current_year - data1['year']
```

```python
# Visualization Car details 
```

```python
# Selling price vs car  age 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=data1, x='car_age', y='selling_price', hue='fuel')
plt.title('Selling Price vs Car Age by Fuel Type')
plt.xlabel('Car Age (years)')
plt.ylabel('Selling Price (INR)')
plt.show()
```

```python
# Distribution of owners
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=data1, x='ower_clean', order=data1['ower_clean'].value_counts().index)
plt.title('Ownership Status Distribution')
plt.xlabel('Owner Type')
plt.ylabel('Count')
plt.show()
```

```python
# Ownership Impact on Price
```

```python
sns.boxplot(data=data1, x='ower_clean', y='selling_price')
plt.title("Price Distribution by Ownership Type")
plt.xlabel("Ownership")
plt.ylabel("Selling Price (INR)")
plt.show()
```

```python
# Top Brands & Models
```

```python
top_brands = data1['brand'].value_counts().head(10)
sns.barplot(x=top_brands.index, y=top_brands.values)
plt.title("Top 10 Most Common Car Brands")
plt.ylabel("Number of Listings")
plt.xlabel("Brand")
plt.xticks(rotation=45)
plt.show()
```

```python
#Mileage vs Price Trade-off
```

```python
sns.scatterplot(data=data1, x='mileage', y='selling_price')
plt.title("Mileage vs Selling Price")
plt.xlabel("Mileage (kmpl)")
plt.ylabel("Selling Price (INR)")
plt.show()
```

```python
#Fuel Type Distribution
```

```python
fuel_counts = data1['fuel'].value_counts()
sns.barplot(x=fuel_counts.index, y=fuel_counts.values)
plt.title("Fuel Type Distribution")
plt.ylabel("Count")
plt.xlabel("Fuel Type")
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(10,6))
sns.heatmap(data1[['selling_price', 'km_driven', 'mileage', 'engine', 'max_power', 'car_age']].corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Matrix')
plt.show()
```
